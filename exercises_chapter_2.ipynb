```python
import os
import tarfile
import urllib
import pandas as pd
```

```python
PATH = os.path.join("datasets","earthquakes")

def load_data(path=PATH):
    csv_path = os.path.join(path, "database.csv")
    return pd.read_csv(csv_path)

earthquakes = load_data()
earthquakes
```

```python
earthquakes_no_NAN= earthquakes.drop(['Depth Error','Depth Seismic Stations', 'Magnitude Error',
                                    'Magnitude Seismic Stations','Azimuthal Gap' ,'Horizontal Distance',
                                    'Horizontal Error'],axis=1)
earthquakes_no_NAN= earthquakes_no_NAN.dropna(subset=["Magnitude"])
earthquakes_no_NAN
```

```python
earthquakes_no_NAN['Date'] = pd.to_datetime(earthquakes_no_NAN['Date'], format="mixed", utc= True)
earthquakes_no_NAN['Date_float'] = earthquakes_no_NAN['Date'].dt.strftime('%m%d%Y').astype(float)
earthquakes_no_NAN['Date_float']
```

```python
earthquakes_no_NAN['Time'] = pd.to_datetime(earthquakes_no_NAN['Time'], format='mixed')
def time_to_seconds(t):
    return t.hour * 3600 + t.minute * 60 + t.second
earthquakes_no_NAN['TimeFloat'] = earthquakes_no_NAN['Time'].apply(time_to_seconds)
```

```python
import matplotlib.pyplot as plt
earthquakes_no_NAN.hist(bins=50, figsize=(20,15))
plt.show
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
import numpy as np

earthquakes_no_NAN["Magnitude"] = pd.cut(earthquakes_no_NAN["Magnitude"],
                               bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                               labels=[1, 2, 3, 4, 5])

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(earthquakes_no_NAN, earthquakes_no_NAN["Magnitude"]):
    strat_train_set = earthquakes_no_NAN.loc[train_index]
    strat_test_set = earthquakes_no_NAN.loc[test_index]
```

```python
earthquakes_no_NAN = strat_train_set.drop("Date_float", axis=1) # drop labels for training set
earthquakes_labels = strat_train_set["Date_float"].copy()
```

```python
object_columns= earthquakes_no_NAN.select_dtypes(include=['object']).columns
earthquakes_num= earthquakes_no_NAN.drop(object_columns, axis=1)
earthquakes_num= earthquakes_num.drop(['Date'],axis=1)

num_columns= earthquakes_no_NAN.select_dtypes(include=['float']).columns
earthquakes_obj=earthquakes_no_NAN.drop(num_columns, axis=1)
earthquakes_obj= earthquakes_obj.drop(['Date','Time','TimeFloat'], axis=1)
```

```python

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

cat_pipeline = Pipeline([
    ('encoder', OneHotEncoder())
])
earthquakes_obj_tr= cat_pipeline.fit_transform(earthquakes_obj)
earthquakes_obj_tr= pd.DataFrame(earthquakes_obj_tr)

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
earthquakes_num_tr= num_pipeline.fit_transform(earthquakes_num)
earthquakes_num_tr= pd.DataFrame(earthquakes_num_tr)
earthquakes_prepared =pd.concat([earthquakes_num, earthquakes_obj], axis=1)
```

```python
from sklearn.compose import ColumnTransformer

earthquakes_prepared =pd.concat([earthquakes_num, earthquakes_obj], axis=1)
num_attribs = list(earthquakes_num_tr)
cat_attribs = list(earthquakes_obj_tr)

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", cat_pipeline, cat_attribs),
])
earthquakes_prepared = full_pipeline.fit_transform(earthquakes_prepared)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
linreg = LinearRegression()
linreg.fit(earthquakes_prepared, earthquakes_labels)
predictions = linreg.predict(earthquakes_prepared)
lin_mse = mean_squared_error(earthquakes_labels, predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(earthquakes_prepared, earthquakes_labels)
housing_predictions = tree_reg.predict(earthquakes_prepared)
tree_mse = mean_squared_error(earthquakes_labels, housing_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
scores = cross_val_score(tree_reg, earthquakes_prepared, earthquakes_labels,
scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
display_scores(-scores)
```

```python
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor()
scoress= cross_val_score(RFR, earthquakes_prepared, earthquakes_labels, scoring='neg_mean_squared_error', cv=10)
rmse =np.sqrt(-scoress)
rmse
display_scores(-scoress)
```

```python
from sklearn.svm import SVR
svr = SVR(kernel= 'rbf')
scoresss= cross_val_score(svr, earthquakes_prepared, earthquakes_labels, scoring='neg_mean_squared_error', cv=10)
rmse2 = np.sqrt(-scoresss)
rmse2
display_scores(-scoresss)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_distribs = {
        'n_estimators': randint(low=1, high=200),
        'max_features': randint(low=1, high=8),
    }

forest_reg = RandomForestRegressor(random_state=42)
rnd_search = RandomizedSearchCV(forest_reg, param_distributions=param_distribs,
                                n_iter=3, cv=5, scoring='neg_mean_squared_error', random_state=42)
rnd_search.fit(earthquakes_prepared, earthquakes_labels)

feature_importance = rnd_search.best_estimator_.feature_importances_
feature_importance
```

```python
# evaluating on test set
final_model = rnd_search.best_estimator_

X_test = strat_test_set.drop("Date_float", axis=1)
y_test = strat_test_set["Date_float"].copy()

X_test_prepared = full_pipeline.transform(X_test)
final_predictions = final_model.predict(X_test_prepared)

final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
final_rmse
```
